```python
# 1. Buat functionnya untuk memetakan data yang diinginkan
# 0 -> False
# 1 -> True

def convert_Legendary(row):
    if row['Legendary'] == False:
        return '0'
    elif row['Legendary'] == True:
        return '1'
    else:
        return 'Unknown'
    

# 2. Apply functionnya

data_1['Legendary_Status'] = data_1.apply(convert_Legendary, axis=1)
```

### Query 1

```python
data_group = data_1.groupby (['Type 1'], as_index = False) ['Legendary_Status'].count ()\
                   .rename (columns = {'Legendary_Status' : 'Total'})\
                   .sort_values ('Total', ascending = False)
```

```python
data_group
```

```python
sns.barplot (data = data_group, x = 'Type 1', y = 'Total')
plt.xlabel ('Type 1')
plt.ylabel ('Total')
plt.xticks (rotation = 45)
plt.show ()
```

### Query 2

```python
query = """
        SELECT 
            *
        FROM
            data_1
        WHERE
            Attack > Defense
        GROUP BY
            Name
        """
result = pandasql.sqldf(query)

result   
```

```python
Attack = result.groupby ('Type 1', as_index = False) ['Legendary_Status'].count ()\
               .rename (columns = {'Legendary_Status' : 'Total'})\
               .sort_values ('Total', ascending = False)
```

### Query 5

```python
# 1. Buat functionnya untuk memetakan data yang diinginkan
# 0 -> False
# 1 -> True

def convert_Type(row):
    if row['Type 1'] == row ['Type 2']:
        return '1'
    elif row['Type 1'] != row ['Type 2']:
        return '0'
    else:
        return 'UNKNOWN'
    

# 2. Apply functionnya

data_1['Dual_Type'] = data_1.apply(convert_Type, axis=1)
```
