**GROUPING AND AGGREGATING**
10. Melakukan sorting values"False" dengan menampilkan total melalui nama kolom "Pokemon Name"

```python
dat.groupby(["Pokemon Name"], sort=False)["Total"].sum()
```

11. Melakukan grouping by Speed (Kecepatan Maksimal)

```python
dat.groupby('Pokemon Name').Speed.agg(
    Speed_Max =('max'),
)
```

12. Melakukan Grouping antara nama Kolom Pokemon Name, Type 1 dan Speed.

```python
dat.groupby(['Pokemon Name','Type 1'])['Speed'].sum()
```

13. Menggunakan Reset Index berfungsi untuk Menyediakan indeks baru untuk dikelompokkan berdasarkan kerangka data dan menjadikannya struktur kerangka data yang tepat.

```python
dat.groupby(['Pokemon Name','Type 2'])['Speed'].sum().reset_index()
```

```python
dat.groupby(['Pokemon Name','Type 1'])['Speed'].sum().reset_index()
```
